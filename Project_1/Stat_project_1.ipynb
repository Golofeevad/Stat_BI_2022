## 1. Read data
Function `read_data` meges datasets from files of specialised type. In our case datasets are in files `.csv`.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
````

````python
import os
def read_data(file_type, folder):
  files_names = []
  for file in os.listdir(folder):
    if file.endswith(file_type):
      files_names.append(os.path.join(folder, file))
  data_table = pd.concat(map(lambda x: pd.read_table(x, sep=',') , files_names), ignore_index=True)
  return data_table
        
````

##14. Height, weight and age of Olimpic games partisipants.##

````python
winter_games = corrected_data[corrected_data['Season']=='Winter']
height_women = winter_games[winter_games['Sex']=='F'].Height.dropna()
weight_women = winter_games[winter_games['Sex']=='F'].Weight.dropna()
age_women = winter_games[winter_games['Sex']=='F'].Age.dropna()
height_men = winter_games[winter_games['Sex']=='M'].Height.dropna()
weight_men = winter_games[winter_games['Sex']=='M'].Weight.dropna()
age_men = winter_games[winter_games['Sex']=='M'].Age.dropna()
````

````python
def find_ci(sample):
    std = sample.std()
    mean_sample = np.mean(sample)
    se = std / np.sqrt(len(sample))
    left = mean_sample - 1.96 * se
    right = mean_sample + 1.96 * se
    return f'mean = {round(mean_sample, 2)}; conf.int. = ({round(left, 2)}-{round(right, 2)})'
````

````python
height_women_interval = find_ci(height_women)
weight_women_interval = find_ci(weight_women)
age_women_interval = find_ci(age_women)
height_men_interval = find_ci(height_men)
weight_men_interval = find_ci(weight_men)
age_men_interval = find_ci(age_men)
````

````python
data_for_publ = {'Mean height' : [height_men_interval, height_women_interval, round(height_men.mean() - height_women.mean(),2)],
                 'Mean weight' : [weight_men_interval, weight_women_interval, round(weight_men.mean() - weight_women.mean(),2)],
                 'Mean age' : [age_men_interval, age_women_interval, round(age_men.mean() - age_women.mean(),2)]}

table_for_publ = pd.DataFrame(data_for_publ, index=['Men', 'Women', 'Difference'])

table_for_publ
````

Everywhere (for hight, weight and age) we have no intersections of confidence interval for men and women. Mean values and difference of this values for men and women represented in table above. It's possible to copy this table to paper or export excel:
